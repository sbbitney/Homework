```python
latitude = np.random.uniform(low = -90.000, high = 90.000, size = 50)
longitude = np.random.uniform(low = -180.000, high = 180.000, size = 50)

city_list = []
for i in range(len(latitude)):
    lat = latitude[i]
    long = longitude[i]
    city = citipy.nearest_city(lat, long).city_name
    country = citipy.nearest_city(lat, long).country_code
    #if not on city_list:
    city_list.append((city, country))

API = 'd791bbcdf03e043bbd2316f92982a22b'
URL = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + API
city_url = URL + "&q=" + city[0] + ',' + city[1]





weather = requests.get(city_url).json()
print(weather)
```
